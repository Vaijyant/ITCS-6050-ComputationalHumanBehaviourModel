# EDA with pyAudioAnalysis

```python
from pyAudioAnalysis import audioBasicIO
from pyAudioAnalysis import audioFeatureExtraction
import matplotlib.pyplot as plt
```

## Feature vs Frame Number for different emotions
(based on short term features)

for the sentence: "*kids are talking by the door*"


```python
[Fs, x] = audioBasicIO.readAudioFile(r"C:\Users\vaijy\Desktop\data\RAVDESS\angry\03-01-05-01-01-01-01.wav")
F = audioFeatureExtraction.stFeatureExtraction(x, Fs, 0.50*Fs, 0.025*Fs)

[Fs1, x1] = audioBasicIO.readAudioFile(r"C:\Users\vaijy\Desktop\data\RAVDESS\calm\03-01-02-01-01-01-01.wav")
F1 = audioFeatureExtraction.stFeatureExtraction(x1, Fs1, 0.50*Fs1, 0.025*Fs1)

[Fs2, x2] = audioBasicIO.readAudioFile(r"C:\Users\vaijy\Desktop\data\RAVDESS\disgust\03-01-07-01-01-01-01.wav")
F2 = audioFeatureExtraction.stFeatureExtraction(x2, Fs2, 0.50*Fs2, 0.025*Fs2)

[Fs3, x3] = audioBasicIO.readAudioFile(r"C:\Users\vaijy\Desktop\data\RAVDESS\fearful\03-01-06-01-01-01-01.wav")
F3 = audioFeatureExtraction.stFeatureExtraction(x3, Fs3, 0.50*Fs3, 0.025*Fs3)

[Fs4, x4] = audioBasicIO.readAudioFile(r"C:\Users\vaijy\Desktop\data\RAVDESS\happy\03-01-03-01-01-01-01.wav")
F4 = audioFeatureExtraction.stFeatureExtraction(x4, Fs4, 0.50*Fs4, 0.025*Fs4)

[Fs5, x5] = audioBasicIO.readAudioFile(r"C:\Users\vaijy\Desktop\data\RAVDESS\neutral\03-01-01-01-01-01-01.wav")
F5 = audioFeatureExtraction.stFeatureExtraction(x5, Fs5, 0.50*Fs5, 0.025*Fs5)

[Fs6, x6] = audioBasicIO.readAudioFile(r"C:\Users\vaijy\Desktop\data\RAVDESS\sad\03-01-04-01-01-01-01.wav")
F6 = audioFeatureExtraction.stFeatureExtraction(x6, Fs6, 0.50*Fs6, 0.025*Fs6)

[Fs7, x7] = audioBasicIO.readAudioFile(r"C:\Users\vaijy\Desktop\data\RAVDESS\surprised\03-01-08-01-01-01-01.wav")
F7 = audioFeatureExtraction.stFeatureExtraction(x7, Fs7, 0.50*Fs7, 0.025*Fs7)
```

```python
plt.rcParams["figure.figsize"] = (10,5)

plt.subplot(2,1,1); plt.plot(F[0,:], label='angry'); plt.xlabel('Frame no'); plt.ylabel('ZCR');
plt.plot(F1[0,:], label='calm')
plt.plot(F2[0,:], label='disgust')
plt.plot(F3[0,:], label='fearful')
plt.plot(F4[0,:], label='happy')
plt.plot(F5[0,:], label='neutral')
plt.plot(F6[0,:], label='sad')
plt.plot(F7[0,:], label='surprised')
plt.legend()
plt.show()

plt.subplot(2,1,1); plt.plot(F[1,:], label='angry'); plt.xlabel('Frame no'); plt.ylabel('Energy');
plt.plot(F1[1,:], label='calm')
plt.plot(F2[1,:], label='disgust')
plt.plot(F3[1,:], label='fearful')
plt.plot(F4[1,:], label='happy')
plt.plot(F5[1,:], label='neutral')
plt.plot(F6[1,:], label='sad')
plt.plot(F7[1,:], label='surprised')
plt.legend()
plt.show()

plt.subplot(2,1,1); plt.plot(F[2,:], label='angry'); plt.xlabel('Frame no'); plt.ylabel('Entropy of Energy');
plt.plot(F1[2,:], label='calm')
plt.plot(F2[2,:], label='disgust')
plt.plot(F3[2,:], label='fearful')
plt.plot(F4[2,:], label='happy')
plt.plot(F5[2,:], label='neutral')
plt.plot(F6[2,:], label='sad')
plt.plot(F7[2,:], label='surprised')
plt.legend()
plt.show()

plt.subplot(2,1,1); plt.plot(F[3,:], label='angry'); plt.xlabel('Frame no'); plt.ylabel('Spectral Centroid');
plt.plot(F1[3,:], label='calm')
plt.plot(F2[3,:], label='disgust')
plt.plot(F3[3,:], label='fearful')
plt.plot(F4[3,:], label='happy')
plt.plot(F5[3,:], label='neutral')
plt.plot(F6[3,:], label='sad')
plt.plot(F7[3,:], label='surprised')
plt.legend()
plt.show()

plt.subplot(2,1,1); plt.plot(F[4,:], label='angry'); plt.xlabel('Frame no'); plt.ylabel('Spectral Spread');
plt.plot(F1[4,:], label='calm')
plt.plot(F2[4,:], label='disgust')
plt.plot(F3[4,:], label='fearful')
plt.plot(F4[4,:], label='happy')
plt.plot(F5[4,:], label='neutral')
plt.plot(F6[4,:], label='sad')
plt.plot(F7[4,:], label='surprised')
plt.legend()
plt.show()

plt.subplot(2,1,1); plt.plot(F[5,:], label='angry'); plt.xlabel('Frame no'); plt.ylabel('Spectral Entropy');
plt.plot(F1[5,:], label='calm')
plt.plot(F2[5,:], label='disgust')
plt.plot(F3[5,:], label='fearful')
plt.plot(F4[5,:], label='happy')
plt.plot(F5[5,:], label='neutral')
plt.plot(F6[5,:], label='sad')
plt.plot(F7[5,:], label='surprised')
plt.legend()
plt.show()

plt.subplot(2,1,1); plt.plot(F[6,:], label='angry'); plt.xlabel('Frame no'); plt.ylabel('Spectral Flux');
plt.plot(F1[6,:], label='calm')
plt.plot(F2[6,:], label='disgust')
plt.plot(F3[6,:], label='fearful')
plt.plot(F4[6,:], label='happy')
plt.plot(F5[6,:], label='neutral')
plt.plot(F6[6,:], label='sad')
plt.plot(F7[6,:], label='surprised')
plt.legend()
plt.show()

plt.subplot(2,1,1); plt.plot(F[7,:], label='angry'); plt.xlabel('Frame no'); plt.ylabel('Spectral Rollof');
plt.plot(F1[7,:], label='calm')
plt.plot(F2[7,:], label='disgust')
plt.plot(F3[7,:], label='fearful')
plt.plot(F4[7,:], label='happy')
plt.plot(F5[7,:], label='neutral')
plt.plot(F6[7,:], label='sad')
plt.plot(F7[7,:], label='surprised')
plt.legend()
plt.show()

plt.subplot(2,1,1); plt.plot(F[33,:], label='angry'); plt.xlabel('Frame no'); plt.ylabel('Chroma Deviation');
plt.plot(F1[33,:], label='calm')
plt.plot(F2[33,:], label='disgust')
plt.plot(F3[33,:], label='fearful')
plt.plot(F4[33,:], label='happy')
plt.plot(F5[33,:], label='neutral')
plt.plot(F6[33,:], label='sad')
plt.plot(F7[33,:], label='surprised')
plt.legend()
plt.show()
```
